```python
# Utility function to move the midpoint of a colormap to be around
# the values of interest.
from matplotlib.colors import Normalize

class MidpointNormalize(Normalize):

    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))
```

```python
def plotHeatmapResults(inputfile,scores='test',title='Heatmap',colorbar='individual'):
    '''
    Arguments
        # colorbar: if 'norm': gives the same colorbar across subjects (10% +- chance)
                    if 'individual': adapts to the specific subject
    
    '''
    C_vals, gamma_vals, train_scores, test_scores = getVals2(inputfile)
    tuple_vals = list(zip([C_vals,gamma_vals,train_scores,test_scores]))       

    if scores == 'test':
        reshape_scores = np.reshape(test_scores,[10,10]) #Chance dimensions here depending on the no. of CV folds
        
    if scores == 'train':
        reshape_scores = np.reshape(train_scores,[10,10])
        
    random_chance = 0.565
    C_range = C_vals[0:10] # Change CV fold here
    gamma_range = gamma_vals[0::10]

    plt.figure(figsize=(8, 6))
    plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
    
    if scores == 'test':
        if colorbar == 'individual':
            plt.imshow(reshape_scores, interpolation='nearest', cmap=plt.cm.RdBu_r)
        if colorbar == 'norm':
            plt.imshow(reshape_scores, interpolation='nearest', cmap=plt.cm.RdBu_r,
               norm=MidpointNormalize(vmin=random_chance - 0.06, midpoint=random_chance),vmax=random_chance+0.06)
            
    if scores == 'train':
        if colorbar == 'individual':
            plt.imshow(reshape_scores, interpolation='nearest', cmap=plt.cm.hot)
        if colorbar == 'norm':
            plt.imshow(reshape_scores, interpolation='nearest', cmap=plt.cm.hot,
               norm=MidpointNormalize(vmin=random_chance - 0.435, midpoint=random_chance),vmax=random_chance + 0.435)
        
    plt.xlabel('C')
    plt.ylabel('gamma')
    plt.colorbar()
    plt.yticks(np.arange(len(gamma_range)), gamma_range)
    plt.xticks(np.arange(len(C_range)), C_range, rotation=45)
    plt.title(title)
    save_str = title[8:10]
    os.chdir('C:\\Users\\Greta\\Documents\\GitHub\\decoding\\results\\cv_round2\\newcolor')
    plt.savefig('CV2_norm_colormap_' + str(save_str) + '.png',bbox_inches = 'tight',dpi=180)
    plt.show()
    
    ## Output max test accuracy ## 
    #testac = tuple_vals[3][0]
    min_index, min_value = min(enumerate(test_scores), key=operator.itemgetter(1))
    max_index, max_value = max(enumerate(test_scores), key=operator.itemgetter(1))
    print('Max test accuracy: ' + str(max_value))
    
    # Find corresponding c and gamma value for max test accuracy
    max_c = C_vals[max_index]
    max_gamma = gamma_vals[max_index]
    print('Optimal C value: ' + str(max_c))
    print('Optimal gamma value: ' + str(max_gamma))
    
    print(plt)
    
    #return tuple_vals
    
    # For returning test_scores in np array
    test_array = np.asarray(test_scores)
    
    return test_array
  
```
